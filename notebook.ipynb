# Imports and functions

```python
from src.models.pred_model import pred_model
from src.models.test_model import test_model
import os
import nibabel as nib
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
from matplotlib import colors
import numpy as np
from src.models.test_model import test_model
import torch
import json

def visualize_image_and_mask(image_data, mask_data):
    """
    Visualize the image and mask, allowing to slide through slices.

    Parameters:
    - image_data: NumPy array of shape (H, W, num_slices)
    - mask_data: NumPy array of shape (H, W, num_slices)
    """
    cmap = colors.ListedColormap(['black', 'red', 'yellow'])  
    bounds = [0, 1, 2, 3]
    norm = colors.BoundaryNorm(bounds, cmap.N)
    
    %matplotlib qt

    num_slices = image_data.shape[2]
    slice_index = 0  

    # Create the figure and axes
    fig, ax = plt.subplots()
    plt.subplots_adjust(bottom=0.25)

    # Display the initial image and mask
    img_slice = image_data[:, :, slice_index]
    mask_slice = mask_data[:, :, slice_index]

    # Display the image slice
    im = ax.imshow(img_slice.T, cmap='gray', origin='lower')
    # Overlay the mask slice with transparency
    msk = ax.imshow(mask_slice, cmap=cmap, norm=norm, alpha=0.5, origin='lower')

    ax.set_title(f'Slice {slice_index + 1}/{num_slices}')

    # Create a slider axis and slider
    ax_slice = plt.axes([0.25, 0.1, 0.65, 0.03])
    slice_slider = Slider(ax_slice, 'Slice', 1, num_slices, valinit=slice_index + 1, valfmt='%0.0f')

    # Update function
    def update(val):
        slice_idx = int(slice_slider.val) - 1  # Adjust for zero-based index
        img_slice = image_data[:, :, slice_idx]
        mask_slice = mask_data[:, :, slice_idx]
        im.set_data(img_slice.T)
        msk.set_data(mask_slice.T)
        ax.set_title(f'Slice {slice_idx + 1}/{num_slices}')
        fig.canvas.draw_idle()

    # Connect the slider to the update function
    slice_slider.on_changed(update)

    plt.show()

def get_image_mask(pred, test_dir):
    test_id = pred[0][0]  

    image_path = os.path.join(test_dir, test_id)

    img = nib.load(image_path)

    img_data = img.get_fdata().astype(np.float32)  
    predicted_masks = pred[1].numpy()
    predicted_masks = np.transpose(predicted_masks, (1, 2, 0))
    
    return img_data, predicted_masks

datasets_dir = './download/dataset'
output_dir = './output'
```

```python
def plot_image_and_mask_side_by_side(image_path, mask_data, slice_index=None):
    """
    Plots two images side by side: one without the mask and one with the mask overlay.
    
    Parameters:
    - image_path: Path to the original image file (without the mask).
    - mask_data: Torch tensor containing the predicted mask data.
    - slice_index: Index of the slice to display. If None, defaults to the middle slice.
    """
    # Load the original image data
    image_data = nib.load(image_path).get_fdata()
    
    # Normalize the image data for consistent visualization
    image_data = (image_data - image_data.min()) / (image_data.max() - image_data.min())
    
    # Determine the slice index (default to middle slice if not provided)
    if slice_index is None:
        slice_index = image_data.shape[2] // 2  # Select middle slice along the third dimension

    # Extract the image and mask slices for the given slice index and rotate by 90 degrees
    image_slice = np.rot90(image_data[:, :, slice_index])
    mask_slice = np.rot90(mask_data[slice_index, :, :].cpu().numpy())  # Convert mask to NumPy array

    # Use different colors for liver (1) and tumor (2)
    cmap = colors.ListedColormap(['black', 'red', 'yellow'])  # Black for background, Red for liver, Yellow for tumor
    norm = colors.BoundaryNorm([0, 1, 2, 3], cmap.N)

    # Plotting
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))

    plt.subplots_adjust(wspace=0.02) 
    
    # Original Image Slice
    axs[0].imshow(image_slice, cmap="gray")
    axs[0].axis("off")
    
    # Image Slice with Mask Overlay
    axs[1].imshow(image_slice, cmap="gray")
    axs[1].imshow(mask_slice, cmap=cmap, norm=norm, alpha=0.4)  # Adjust alpha for transparency
    axs[1].axis("off")
    
    plt.show()

```

```python
liver_image_id = pred_liver[0][0]
image_path = os.path.join(liver_test_dir, liver_image_id)
mask_data = pred_liver[1]

# Plot side by side for the middle slice
plot_image_and_mask_side_by_side(image_path, mask_data, slice_index=180)
```
