```python
import pandas as pd
# List to store details of each file
file_details = []

# Loop through each file in the directory and get its metadata
for file_name in os.listdir(image_dir):
    if file_name.endswith('.nii.gz') and not file_name.startswith('._'):
        file_path = os.path.join(image_dir, file_name)
        
        # Load image and get metadata
        image = nib.load(file_path)
        image_data = image.get_fdata()
        shape = image_data.shape  # Image dimensions
        dtype = image_data.dtype  # Data type (e.g., float32, int16)
        voxel_dims = image.header.get_zooms()  # Voxel dimensions (spacing)

        # Append details to the list
        file_details.append({
            "File Name": file_name,
            "Shape": shape,
            "Data Type": str(dtype),
            "Voxel Dimensions": voxel_dims
        })

# Convert the details into a DataFrame and display unique values
file_df = pd.DataFrame(file_details)
unique_values = file_df.nunique()
file_df
```

```python
# Assuming 'file_df' is already defined
pd.set_option('display.max_rows', None)
shape_list = file_df['Shape']
shapes = []
count_threshold_pass = 0
threshold = 300  # Set your threshold value

for shape in shape_list:
    shapes.append(shape[2])
    if shape[2] < threshold:
        count_threshold_pass += 1

# Plot histogram
plt.hist(shapes, bins=20, color='skyblue', edgecolor='black')
plt.axvline(threshold, color='red', linestyle='--', label=f'Threshold = {threshold}')

# Title and labels
plt.title('Distribution of Liver Images Shape Dimension Values')
plt.xlabel('Shape Dimension')
plt.ylabel('Frequency')
plt.legend()

# Display counts
print("Count below threshold:", count_threshold_pass)
print("Total shapes:", len(shape_list))

plt.show()
```
